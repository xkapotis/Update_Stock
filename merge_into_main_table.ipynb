```
table_from_parquet_file = spark.sql("""
select distinct
first(UpdateID) as UpdateID,
first(AvailableQuantity) as AvailableQuantity,
first(AvailableQuantityActual) as AvailableQuantityActual,
SKU,
StoreId,
max(offset) as offset
from 
source_tables_rt.table_from_parquet_file
group by sku, storeid 
""") 
table_from_parquet_file.createOrReplaceTempView("table_from_parquet_file")
```

```
%sql
MERGE INTO source_tables_rt.stock as main
USING table_from_parquet_file as new
ON main.SKU = new.SKU AND main.storeId = new.storeId
WHEN MATCHED 
  THEN UPDATE SET main.UpdateID = new.UpdateID, main.AvailableQuantity = new.AvailableQuantity, main.offSet = new.offSet 
WHEN NOT MATCHED
  THEN INSERT (main.UpdateID, main.AvailableQuantity,main.SKU,main.StoreId,main.offset) VALUES (new.UpdateID, new.AvailableQuantity,new.SKU,new.StoreId,new.offset)
```
